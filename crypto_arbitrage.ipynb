#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

*NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
def arbitrage_calc(exch1_sliced, exch2_sliced, col, date):
    """
    Perform arbitrage calcuations on two exchange trade dataframes,
    generating spread, spread_return, profitable_trades, profit_per_trade, cumulative_profit
    
    Parameters: 
    exch1_sliced, exch2_sliced: Exchange trade data sliced to only a Timestamp index and Close column.
    col: string, column of interest (so this should (?) work even if the exchange trade data was not already sliced)
    date: string, of the form "YYYY-MM-DD"
    
    DataFrames returned: spread, returns, profitable_trades, profit_per_trade, cumulative_profit
    """
    # Measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one
    spread = exch1_sliced[col].loc[date] - exch2_sliced[col].loc[date]
    # Remove any negative spreads
    spread = spread[spread > 0]

    returns = spread / exch2_sliced[col].loc[date]

    # determine the number of times your trades with positive returns 
    # exceed the 1% minimum threshold (.01) that you need to cover your costs
    profitable_trades = returns[returns > 0.01]

    # calculate the potential profit per trade in dollars 
    # Multiply the profitable trades by the cost of the Bitcoin that was purchased
    profit = profitable_trades * exch2_sliced[col].loc[date]
    
    # Drop any missing values from the profit DataFrame
    profit_per_trade = profit.dropna()
    
    # Use the cumsum function to calculate the cumulative profits over time for the early profit per trade DataFrame
    cumulative_profit = profit_per_trade.cumsum()
    
    return spread, returns, profitable_trades, profit_per_trade, cumulative_profit

arbitrage_spread_early, spread_return_early, profitable_trades_early, profit_per_trade_early, cumulative_profit_early \
  = arbitrage_calc(coinbase_sliced, bitstamp_sliced, "Close", "2018-01-16")

arbitrage_spread_middle, spread_return_middle, profitable_trades_middle, profit_per_trade_middle, cumulative_profit_middle \
  = arbitrage_calc(coinbase_sliced, bitstamp_sliced, "Close", "2018-02-24")

arbitrage_spread_late, spread_return_late, profitable_trades_late, profit_per_trade_late, cumulative_profit_late \
  = arbitrage_calc(coinbase_sliced, bitstamp_sliced, "Close", "2018-03-26")

print("Arbitrage spread summaries")
print("\n2018-01-16:\n", arbitrage_spread_early.describe())
print("\n2018-02-24:\n", arbitrage_spread_middle.describe())
print("\n2018-03-26:\n", arbitrage_spread_late.describe())
```
